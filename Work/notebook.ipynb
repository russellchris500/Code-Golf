```python
def p(g,E=enumerate,L=len):
  n=0
  while 1:
    P=[(i,j)for i,r in E(h)for j,v in E(r)if v and
       ((i and h[i-1][j]==v)+(j and r[j-1]==v)+(i+1<L(h)and h[i+1][j]==v)+(j+1<L(r)and r[j+1]==v))==1]
    if not P:break
    R=set();C=set()
    for i,j in P:h[i][j]=0;n+=1;R.add(i);C.add(j)
    for i in sorted(R,reverse=True):
      if all(x==0 for x in h[i]):h.pop(i)
    for j in sorted(C,reverse=True):
      if h and all(r[j]==0 for r in h):
        for r in h:r.pop(j)
  return g
```
